## Part 3) Loading the Data

In the following cell write the code necessary to acquire the data from your online source.
**Do not copy the data itself into this notebook.**

```python

# Install KaggleHub package if not already available
%pip install kagglehub
import os
import kagglehub
from pathlib import Path


def download_dataset(dataset_slug: str, download_dir: str = "") -> str:
    """
    Downloads a dataset from Kaggle using KaggleHub and returns the download path.
    
    Args:
        dataset_slug (str): The dataset identifier in format 'username/dataset-name'.
        download_dir (str, optional): Custom directory to download the dataset to. 
                            If empty string, uses default KaggleHub location.

    Returns:
        str: Local filesystem path where the dataset files were downloaded
    """
    try:
        if download_dir:
            download_dir = str(Path(download_dir).absolute())
            os.makedirs(download_dir, exist_ok=True)
            os.environ['KAGGLE_HUB_DATASETS_DIR'] = download_dir

        path = kagglehub.dataset_download(dataset_slug)
        return path
    
    except OSError as e:
        raise OSError(f"Failed to create/access download directory {download_dir}") from e
    except Exception as e:
        if "Could not connect" in str(e):
            raise ConnectionError("Failed to connect to Kaggle servers") from e
        if "Forbidden" in str(e):
            raise PermissionError("You don't have permission to access this dataset") from e
        if "not found" in str(e).lower():
            raise ValueError("Dataset not found - check the slug format") from e
        raise
```

## Part 4) Cleaning the data

Your data set may have some quirks, strange formatting, incomplete entries or invalid data.
Note down any particularities you find and your intended steps to correct them.
Clean up the raw data you have loaded to bring it into a presentable shape, that can serve as the basis of future processing steps without having to worry about corner cases.
You may create, split, combine, discard, re-format or re-label rows and columns as needed.
Try to keep the cleaning procedure as generic as possible, so it could also work on a different data set of the same kind.

```python
import pandas as pd
import re
from typing import Tuple

def clean_sentiment_data(df: pd.DataFrame) -> Tuple[pd.DataFrame, dict]:
    """
    Cleans and preprocesses sentiment analysis data.
    
    Args:
        df (pd.DataFrame): Raw sentiment analysis dataframe
        
    Returns:
        Tuple[pd.DataFrame, dict]: Cleaned dataframe and dictionary with cleaning statistics
    """
    cleaning_stats = {}
    
    try:
        # Initial stats
        cleaning_stats['initial_rows'] = len(df)
        cleaning_stats['initial_columns'] = df.columns.tolist()
        
        # 1. Remove redundant columns
        if 'Unnamed: 0' in df.columns:
            df.drop('Unnamed: 0', axis=1, inplace=True)
            cleaning_stats['dropped_columns'] = ['Unnamed: 0']
        
        # 2. Handle missing values
        missing_values = df.isnull().sum().to_dict()
        cleaning_stats['missing_values_before'] = missing_values
        
        df['Comment'] = df['Comment'].fillna("")
        cleaning_stats['missing_values_after'] = df.isnull().sum().to_dict()
        
        # 3. Clean text data
        df['Comment'] = df['Comment'].apply(lambda x: re.sub(r'\d+', '', str(x)))  # Remove numbers
        df['Comment'] = df['Comment'].str.lower()  # Convert to lowercase
        df['Comment'] = df['Comment'].str.strip()  # Remove whitespace
        
        # 4. Validate sentiment labels
        unique_sentiments = df['Sentiment'].unique()
        cleaning_stats['unique_sentiments'] = unique_sentiments.tolist()
        
        # 5. Final validation
        assert df['Comment'].isnull().sum() == 0, "Missing values still exist in comments"
        assert len(df) > 0, "All rows were removed during cleaning"
        
        cleaning_stats['final_rows'] = len(df)
        cleaning_stats['final_columns'] = df.columns.tolist()
        
        return df, cleaning_stats
    
    except Exception as e:
        print(f"Error during cleaning: {str(e)}")
        raise


def load_and_clean_dataset(dataset_path: str) -> Tuple[pd.DataFrame, dict]:
    """
    Loads and cleans the dataset from the given path.
    
    Args:
        dataset_path (str): Path to the dataset directory
        
    Returns:
        Tuple[pd.DataFrame, dict]: Cleaned dataframe and cleaning statistics
    """
    try:
        # Find the first CSV file in the directory
        csv_files = list(Path(dataset_path).glob("*.csv"))
        if not csv_files:
            raise FileNotFoundError("No CSV files found in dataset directory")
        
        data_path = csv_files[0]
        print(f"Loading dataset from: {data_path}")
        
        # Try different encodings if needed
        try:
            df = pd.read_csv(data_path, encoding='utf-8')
        except UnicodeDecodeError:
            df = pd.read_csv(data_path, encoding='latin1')
        
        # Clean the data
        cleaned_df, stats = clean_sentiment_data(df)
        return cleaned_df, stats
        
    except Exception as e:
        print(f"Error loading dataset: {str(e)}")
        raise

# Main execution
if __name__ == "__main__":
    try:
        # 1. Download dataset
        custom_download_dir = "D:/datasets/kaggle"
        dataset_path = download_dataset(
            "abdelmalekeladjelet/sentiment-analysis-dataset",
            download_dir=custom_download_dir
        )
        print(f"Dataset downloaded to: {dataset_path}")
        
        # 2. Load and clean
        cleaned_df, cleaning_stats = load_and_clean_dataset(dataset_path)
        
        # 3. Show results
        print("\n=== Cleaning Statistics ===")
        for key, value in cleaning_stats.items():
            print(f"{key}: {value}")
            
        print("\n=== Sample Data ===")
        print(cleaned_df.head())
        
        print("\n=== Dataset Info ===")
        print(f"Final shape: {cleaned_df.shape}")
        print("Unique sentiments:", cleaned_df['Sentiment'].unique())
        
    except Exception as e:
        print(f"Error in main execution: {str(e)}")
```

## Part 5) Fundamental Exploration

Create a statistical analysis for each of the properties recorded in the data set.

* For numerical data calculate the minimum and maximum values and where/when they appear, the mean value as well as the standard derivation.
* For categorical data create a table how often each category appears
* For time/date colums indicate the covered timespan and the average frequency of events
* For any other kind of data discuss and implement a suitable statistical characterization

```python
# Implement your statistical analysis here
# Print the analysis results
def fundamental_exploration(df: pd.DataFrame) -> dict:
    """

    Perform comprehensive statistical analysis of a sentiment analysis dataset.
    
    This function conducts a thorough examination of all data types present in the dataset,
    including numerical, categorical, text, and temporal features. It returns a structured
    dictionary containing all statistical findings.
    
    Analysis includes:
    - Basic dataset metadata (shape, columns, dtypes)
    - Numerical feature statistics (min, max, mean, std) with example records
    - Categorical feature distributions (value counts, unique values)
    - Special handling for sentiment labels (distribution percentages)
    - Text feature analysis (length statistics, word counts, patterns)
    - Temporal analysis for date/time columns (timespan, frequency)
    - Correlation matrix for numerical features
    
    Parameters:
        df (pd.DataFrame): A cleaned pandas DataFrame containing the sentiment analysis data.
                           Expected to have at minimum 'Comment' and 'Sentiment' columns.
    
    Returns:
        dict: A nested dictionary containing all analysis results organized by category:
              {
                  'dataset_shape': tuple,
                  'columns': list,
                  'data_types': dict,
                  'numerical_analysis': {
                      column_name: {
                          'min': float,
                          'max': float,
                          'mean': float,
                          'std': float,
                          'min_example': str,
                          'max_example': str
                      }
                  },
                  'categorical_analysis': {
                      column_name: {
                          'value_counts': dict,
                          'unique_values': int,
                          'most_common': tuple
                      }
                  },
                  'text_analysis': {
                      'avg_length': float,
                      'min_length': int,
                      'max_length': int,
                      'word_counts': dict,
                      'common_prefixes': dict,
                      'common_suffixes': dict
                  },
                  'temporal_analysis': {
                      column_name: {
                          'time_span': dict,
                          'frequency': dict
                      }
                  },
                  'correlation_matrix': dict (if applicable)
              }
    
    Raises:
        TypeError: If input is not a pandas DataFrame
        ValueError: If required columns are missing
        Exception: For other analysis-specific errors
    
    """
    analysis_results = {}
    
    # 1. General Dataset Overview
    analysis_results['dataset_shape'] = df.shape
    analysis_results['columns'] = df.columns.tolist()
    analysis_results['data_types'] = df.dtypes.to_dict()
    
    # 2. Numerical Data Analysis (if any exists)
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    numerical_stats = {}
    
    for col in numerical_cols:
        if col == 'Sentiment':  # Treat sentiment separately as categorical
            continue
            
        col_stats = {
            'min': df[col].min(),
            'max': df[col].max(),
            'mean': df[col].mean(),
            'std': df[col].std(),
            'min_example': df.loc[df[col] == df[col].min(), 'Comment'].iloc[0] if 'Comment' in df.columns and not df.loc[df[col] == df[col].min(), 'Comment'].empty else None,
            'max_example': df.loc[df[col] == df[col].max(), 'Comment'].iloc[0] if 'Comment' in df.columns and not df.loc[df[col] == df[col].max(), 'Comment'].empty else None
        }
        numerical_stats[col] = col_stats
    
    analysis_results['numerical_analysis'] = numerical_stats
    
    # 3. Categorical Data Analysis
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    categorical_stats = {}
    
    for col in categorical_cols:
        value_counts = df[col].value_counts().to_dict()
        categorical_stats[col] = {
            'value_counts': value_counts,
            'unique_values': len(value_counts),
            'most_common': max(value_counts.items(), key=lambda x: x[1])
        }
    
    # Special handling for Sentiment (treated as categorical)
    if 'Sentiment' in df.columns:
        sentiment_stats = {
            'value_counts': df['Sentiment'].value_counts().to_dict(),
            'unique_values': df['Sentiment'].nunique(),
            'distribution_percentage': (df['Sentiment'].value_counts(normalize=True) * 100).to_dict()
        }
        categorical_stats['Sentiment'] = sentiment_stats
    
    analysis_results['categorical_analysis'] = categorical_stats
    
    # 4. Text Data Analysis
    if 'Comment' in df.columns:
        text_stats = {
            'avg_length': df['Comment'].str.len().mean(),
            'min_length': df['Comment'].str.len().min(),
            'max_length': df['Comment'].str.len().max(),
            'word_counts': {
                'avg': df['Comment'].str.split().str.len().mean(),
                'min': df['Comment'].str.split().str.len().min(),
                'max': df['Comment'].str.split().str.len().max()
            },
            'common_prefixes': df['Comment'].str[:3].value_counts().head(5).to_dict(),
            'common_suffixes': df['Comment'].str[-3:].value_counts().head(5).to_dict()
        }
        analysis_results['text_analysis'] = text_stats
    
    # 5. Temporal Analysis (if date/time columns exist)
    datetime_cols = df.select_dtypes(include=['datetime64']).columns
    temporal_stats = {}
    
    for col in datetime_cols:
        col_stats = {
            'time_span': {
                'start': df[col].min(),
                'end': df[col].max(),
                'duration_days': (df[col].max() - df[col].min()).days
            },
            'frequency': {
                'avg_days_between': (df[col].diff().dt.days.mean()),
                'min_days_between': (df[col].diff().dt.days.min()),
                'max_days_between': (df[col].diff().dt.days.max())
            }
        }
        temporal_stats[col] = col_stats
    
    analysis_results['temporal_analysis'] = temporal_stats
    
    # 6. Correlation Analysis (if multiple numerical columns exist)
    if len(numerical_cols) > 1:
        analysis_results['correlation_matrix'] = df[numerical_cols].corr().to_dict()
    
    return analysis_results

# 4. Perform fundamental exploration
print("\n=== Starting Fundamental Exploration ===")
analysis_results = fundamental_exploration(cleaned_df)
        
# 5. Display key analysis results
print("\n=== Fundamental Analysis Results ===")
print(f"\nSentiment Distribution:")
print(analysis_results['categorical_analysis']['Sentiment']['value_counts'])
        
if 'text_analysis' in analysis_results:
    print(f"\nText Statistics:")
    print(f"Average comment length: {analysis_results['text_analysis']['avg_length']:.1f} chars")
    print(f"Average word count: {analysis_results['text_analysis']['word_counts']['avg']:.1f} words")
        
```

```python
# Implement your visualization here
%pip install seaborn
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
def visualize_analysis(results: dict, df: pd.DataFrame):
    """
    Create visualizations based on the analysis results.
    
    Args:
        results (dict): Analysis results from fundamental_exploration
        df (pd.DataFrame): The cleaned dataframe
    """
    # 1. Sentiment Distribution
    if 'Sentiment' in df.columns:
        plt.figure(figsize=(10, 5))
        sns.countplot(x='Sentiment', data=df)
        plt.title('Distribution of Sentiment Labels')
        plt.show()
        
        # Pie chart of sentiment distribution
        sentiment_counts = df['Sentiment'].value_counts()
        plt.figure(figsize=(8, 8))
        plt.pie(sentiment_counts, labels=[str(label) for label in sentiment_counts.index], autopct='%1.1f%%')
        plt.title('Sentiment Label Proportions')
        plt.show()
    
    # 2. Comment Length Distribution
    if 'Comment' in df.columns:
        df['comment_length'] = df['Comment'].str.len()
        plt.figure(figsize=(10, 5))
        sns.histplot(data=df, x='comment_length', bins=30)
        plt.title('Distribution of Comment Lengths')
        plt.xlabel('Number of Characters')
        plt.show()
    
    # 3. Word Count Distribution
    if 'Comment' in df.columns:
        df['word_count'] = df['Comment'].str.split().str.len()
        plt.figure(figsize=(10, 5))
        sns.histplot(data=df, x='word_count', bins=30)
        plt.title('Distribution of Word Counts per Comment')
        plt.xlabel('Number of Words')
        plt.show()
```

```python
# 6. Generate visualizations
print("\nGenerating visualizations...")
visualize_analysis(analysis_results, cleaned_df)
        
```
